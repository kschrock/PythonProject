```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
import seaborn as sns
import array
```

```python
#Read the csv file into a dataframe
df = pd.read_csv('USvideos.csv')
```

```python
df_top100views = pd.DataFrame(df.nlargest(100, 'views'))

```

```python

sns.pairplot(df, vars=['category_id','views','likes'], diag_kind = 'hist',kind = 'scatter', hue = 'category_id')
plt.show()
```

```python
sns.pairplot(df_top100views, vars=['category_id','views','likes'], diag_kind = 'hist',kind = 'scatter', hue = 'category_id')
plt.show()
```

```python
#Looking at both graphs
print('1. We can see that country 24 has the most likes and views consistently in both graphs.')
print('2. We also can see the more views a video has the more likes it will have compared to other videos with less views.')
print('3. We also find that video views tend to drift upwards. This means that trending video views are expected to be higher in 2018 than they were in 2017')
print('4. The top 100 videos have more likes and views than the other videos who didnt make the top 100.')
print('5. Both graphs a similar but the main difference is the top 100 videos are more exponetial.')
```

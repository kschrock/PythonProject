```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('Book.csv')
```

```python
ax1 = df.plot.scatter(x='Leaflets', y='Sales', c='Green')
#This is a scatter-plot chart that shows Leaflets on the X axis and Sales on the Y axis.
```

```python
hist = df.hist(column='Revenue', bins=10)
# This is a histogram that shows Revenue distributed into 10 bins, right skewed
```

```python
sns.set_style("whitegrid") 
  
sns.boxplot(x = 'Location', y = 'Sales', data = df)
#Box and Whisker plot for two Location (one box for Park, one for Beach) showing Total Sales values
```

```python
location = df.loc[df['Temperature']>= 75, 'Location'].values
#Getting only Location values that have a Temperature of 75 or greater
sales = df.loc[df['Temperature']>= 75, 'Sales'].values
#Getting only Sales values that have a Temperature of 75 or greater
```

```python
df2 = pd.DataFrame()
#Create new data frame
df2['location'] = location   
# add the array to df as a column
df2['sales'] = sales
# add the array to df as a column
```

```python
sns.set_style("whitegrid") 
  
sns.boxplot(x = 'location', y = 'sales', data = df2)
#Box and Whisker plot for two Location (one box for Park, one for Beach) showing Total Sales
#values. For only the sales of the day that Temperature was at least 75 F
```

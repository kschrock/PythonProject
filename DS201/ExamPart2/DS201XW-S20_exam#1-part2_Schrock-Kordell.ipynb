```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
import seaborn as sns
import array
```

```python
#Read the csv file into a dataframe
df = pd.read_csv('Automobile_price_data_clean-f18.csv')
```

```python
#This grabs all the car data I want from the csv file
alfaromero = df.loc[df['make'] == 'alfa-romero', 'price'].values
audi= df.loc[df['make'] == 'audi', 'price'].values
bmw= df.loc[df['make'] == 'bmw', 'price'].values
chevrolet= df.loc[df['make'] == 'chevrolet', 'price'].values
dodge= df.loc[df['make'] == 'dodge', 'price'].values
honda= df.loc[df['make'] == 'honda', 'price'].values
isuzu= df.loc[df['make'] == 'isuzu', 'price'].values
jaguar= df.loc[df['make'] == 'jaguar', 'price'].values
mazda= df.loc[df['make'] == 'mazda', 'price'].values
mercedesbenz= df.loc[df['make'] == 'mercedes-benz', 'price'].values
mercury= df.loc[df['make'] == 'mercury', 'price'].values
mitsubishi= df.loc[df['make'] == 'mitsubishi', 'price'].values
nissan= df.loc[df['make'] == 'nissan', 'price'].values
peugot= df.loc[df['make'] == 'peugot', 'price'].values
plymouth= df.loc[df['make'] == 'plymouth', 'price'].values
porsche= df.loc[df['make'] == 'porsche', 'price'].values
renault= df.loc[df['make'] == 'renault', 'price'].values
saab= df.loc[df['make'] == 'saab', 'price'].values
subaru= df.loc[df['make'] == 'subaru', 'price'].values
toyota= df.loc[df['make'] == 'toyota', 'price'].values
volkswagen= df.loc[df['make'] == 'volkswagen', 'price'].values
volvo= df.loc[df['make'] == 'volvo', 'price'].values

#create a average function
def Average(lst): 
    return sum(lst) / len(lst) 

#get the average for each car 
alfaromero = Average(alfaromero)
audi = Average(audi)
bmw = Average(bmw)
chevrolet = Average(chevrolet)
dodge = Average(dodge)
honda = Average(honda)
isuzu = Average(isuzu)
jaguar = Average(jaguar)
mazda = Average(mazda)
mercedesbenz = Average(mercedesbenz)
mercury = Average(mercury)
mitsubishi = Average(mitsubishi)
nissan = Average(nissan)
peugot = Average(peugot)
plymouth = Average(plymouth)
porsche = Average(porsche)
renault = Average(renault)
saab = Average(saab)
subaru = Average(subaru)
toyota = Average(toyota)
volkswagen = Average(volkswagen)
volvo = Average(volvo)

#create x value names
xList = [
'alfa-romero',
'audi',
'bmw',
'chevrolet',
'dodge',
'honda',
'isuzu',
'jaguar',
'mazda',
'mercedes-benz',
'mercury',
'mitsubishi',
'nissan',
'peugot',
'plymouth',
'porsche',
'renault',
'saab',
'subaru',
'toyota',
'volkswagen',
'volvo']
#create y value (price average) for each car 
ylist = [alfaromero,
audi,
bmw,
chevrolet,
dodge,
honda,
isuzu,
jaguar,
mazda,
mercedesbenz,
mercury,
mitsubishi,
nissan,
peugot,
plymouth,
porsche,
renault,
saab,
subaru,
toyota,
volkswagen,
volvo]

plt.xticks(rotation=90)
plt.xlabel('Maker')
plt.ylabel('Average Price')
plt.bar(xList, ylist, align='center',width=.8, color = 'rgbkymc' )
# Question 1, using pandas on jupiter notebook to plot a bar chart “Average of Price by Make”
```

```python
carsAbove8k = df.loc[df['price']>= 8000, 'make'].values
# Question 2, How many toyota cars are priced above USD 8,000? 
count = 0
# Iterate through each car in the array, then increase count if matches toytota
for make in carsAbove8k :
    if(make == 'toyota'):
        count = count +1

print("How many toyota cars are priced above USD 8,000? \n There are " + str(count) + " toyotas above 8k.")
```

```python
hatchbackCarsMPG = df.loc[df['body-style'] == 'hatchback', 'city-mpg'].values
#Question 3, What is the average city-mpg for hatchback cars?
totalMiles = 0
# Iterate through each city-mpg in the hatchback array, then add up the total and average it out.
for mpg in hatchbackCarsMPG :
    totalMiles = totalMiles + mpg
    
averageMPG = totalMiles / len(hatchbackCarsMPG)
print("What is the average city-mpg for hatchback cars? \n The average MPG is " + str(averageMPG) + 
" for hatchback cars.")

```
